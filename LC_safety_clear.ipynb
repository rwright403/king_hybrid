 US Dept of Transportation FAA Calculation of Safety Clear Zones for Experimental Permits under 14 CFR § 437.53(a) 

 See [48]


a) Identify propellant: FROM Table E-2 in Appendix E of 14 CFR part 420. No nitrous paraffin so select 20% which corersponds to Launch Pads LO^2/RP-1 which is the highest TNT equivalence and believed to be conservative

```python
#inputs

m_ox = 29.4835*2 #13.25
m_fuel = 2.7

m_propellant = 2.20462*(m_ox+m_fuel) # [lbm] convert to imperial
print(f"propellant mass: {m_propellant:.3f} [lbm] ")
```

a) net explosive weight:

Referenced in [48],  Appendix E, table E-2 of 14 CFR part 420:
https://www.ecfr.gov/current/title-14/chapter-III/subchapter-C/part-420


```python
yield_factor = 0.2 #see md above
W_net_xplosive = yield_factor*m_propellant

print(f"W_net_xplosive: , {W_net_xplosive} [lbm]")
```

c) compute peak incident overpressure distance

```python
K_fac = 45 # ft/lb^1/3
D_overpressure = K_fac*W_net_xplosive**(1/3)

print(f"D_overpressure: , {D_overpressure} [ft]")
```

d) compute hazardous fragment distance

```python
HFD = 291.3 + (79.2*np.log(W_net_xplosive))

print(f"HFD: , {HFD} [ft] (corresponding to propellant wt under 100lb)")
```

e) set min dimension of safety clear distance to the larger of the hazardous fragment distance (HFD) or the explosive overpressure distance (D).

```python
safety_clear_distance = max(D_overpressure, HFD)
print("safety clear distance: (max of HFD or overpressure) ", safety_clear_distance, "ft")
```
